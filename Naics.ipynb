```python
import pandas as pd
```

```python
rtra_3type1 = pd.read_csv('RTRA_Employ_3NAICS_00_05.csv')
rtra_3type2 = pd.read_csv('RTRA_Employ_3NAICS_06_10.csv')
rtra_3type3 = pd.read_csv('RTRA_Employ_3NAICS_11_15.csv')
rtra_3type4= pd.read_csv('RTRA_Employ_3NAICS_16_20.csv')
rtra_3type5 = pd.read_csv('RTRA_Employ_3NAICS_97_99.csv')
```

```python
lmo_data = pd.read_excel('LMO_Detailed_Industries_by_NAICS.xlsx')
lmo_data['NAICS'] = lmo_data['NAICS'].replace(regex='&', value=',').astype('str')

lmo_data.head()
```

```python
lmo_data.info()
```

```python
two_digit = {}
three_digit = {} 
four_digit = {}

for name, numbers in zip(lmo_data['LMO_Detailed_Industry'], lmo_data['NAICS']):
    num_list = numbers.split(',')
    num_list = [x.strip() for x in num_list]
    for i in range(len(num_list)):
        if len(num_list[i]) == 2:
            two_digit[num_list[i]] = name
        elif len(num_list[i]) == 3:
            three_digit[num_list[i]] = name 
        elif len(num_list[i]) == 4:
            four_digit[num_list[i]] = name
```
